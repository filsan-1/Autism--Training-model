# Autism Detection using CNN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Load Images
# Assuming images are in a directory named 'images'
image_datagen = ImageDataGenerator(rescale=1./255)
batch_size = 32
image_generator = image_datagen.flow_from_directory('images', target_size=(64, 64),
 class_mode='binary', batch_size=batch_size)
```

```python
# Split Data
X_train, X_test, y_train, y_test = train_test_split(image_generator, test_size=0.2, random_state=42)
```

```python
# Build CNN Model
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile Model
model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train Model
history = model.fit(image_generator, epochs=10)
```

```python
# Evaluate Model
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Loss: {loss}, Accuracy: {accuracy}')
```

```python
# Visualize Results
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['loss'], label='loss')
plt.title('Model accuracy and loss')
plt.ylabel('value')
plt.xlabel('epoch')
plt.legend(loc='upper left')
plt.show()
```

```python
# Save Model
model.save('autism_detection_model.h5')
```
